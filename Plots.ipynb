```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("Book1.csv")
```

```python
data
```

# 1) show the total profit of all months using line plot

#### in matplotlib

```python
plt.plot(data[["month_number"]],data[["total_profit"]])
plt.xlabel("month")
plt.ylabel("total profit")
plt.show()
```

#### in seaborn

```python
sns.lineplot(x = "month_number", y = "total_profit", data = data)
```

# 2) show all product sales data using multiline plot

#### in matplotlib

```python
fig, ax = plt.subplots()
plt.figure(figsize=(10,10))
for i in range(1,len(data.columns)-2):
    ax.plot(data[["month_number"]],data.iloc[:,i])
ax.legend(list(data.columns)[1:7], loc = "upper right")
plt.xlabel
plt.show()
```

#### in seaborn

```python
for i in range(1,len(data.columns)-2):
    sns.lineplot(data = data, x = "month_number", y = list(data.columns)[i])
```

# 3) calculate total sales data for each project using pie chart

```python
df = data.iloc[:,1:7].sum()
```

```python
plt.figure(figsize=(10,10))
plt.pie(df, labels = list(data.columns)[1:7], autopct = '%1.1f%%')
```
